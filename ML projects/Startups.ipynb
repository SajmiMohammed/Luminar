```python
# Seperating x & y
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
# Plotting each independent variable with dependent variable.
# R&D vs Profit
import seaborn as sns
sns.regplot(x=df['R&D Spend'],y=y,color='red')
```

```python
# Marketing Spend Vs Profit.
sns.regplot(x=df['Marketing Spend'],y=y,color='green')
```

```python
# Administration Vs Profit
sns.regplot(x=df['Administration'],y=y,color='m')
```

```python
# Onehot encoding(col=>State)
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
col_transfer=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
x=col_transfer.fit_transform(x)
x     # forms 6 columns ,ie state has expanded to 3 columns(florida,newyork,california)
```

```python
# Training & Testing Data.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
# Model creation => Multiple Linear Regression
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
# Creating Dataframe with actual values,predicted values and difference.
values={'actual_value':y_test,'predicted_value':y_pred,'Difference':y_test-y_pred}
df3=pd.DataFrame(values)
df3
```

```python
# Slope
print("Slope is")
list(zip(x,model.coef_))
```

```python
# Constant
print("constant is",model.intercept_)
```

```python
# Performance evaluation.

# a] MAE (mean_absolute_error)
from sklearn.metrics import mean_absolute_error
print("MAE Is ",mean_absolute_error(y_test,y_pred))
```

```python
# b] MAPE (mean_absolute percentage error)
from sklearn.metrics import mean_absolute_percentage_error
print("MAPE is",mean_absolute_percentage_error(y_pred,y_test))
```

```python
# c]MSE (mean_squared_error)
from sklearn.metrics import mean_squared_error
print("MSE Is ",mean_squared_error(y_test,y_pred))
```

```python
# d] RSE (Root_squared_error)
data=mean_squared_error(y_test,y_pred)
print(np.sqrt(data))
```

```python
from sklearn.metrics import r2_score
print("R2 score",r2_score(y_pred,y_test))
```
