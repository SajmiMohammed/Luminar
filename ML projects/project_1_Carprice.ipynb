```python
#Encoding using Get_dummies
df1=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
df1    #contains encoded values only
```

```python
#combining 2 df :use concatenation
dfe=pd.concat([df,df1],axis=1)
dfe
```

```python
dfe=dfe.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price'],axis=1)
dfe
```

```python
#remove units of col ==>mileage,power,engine
# replacing str
dfe['Mileage']=dfe['Mileage'].str.replace('km/kg','')  #replacevalue:km/kg ==>''nothing
dfe['Mileage']=dfe['Mileage'].str.replace('kmpl','')
dfe['Engine']=dfe['Engine'].str.replace('CC','')
dfe['Power']=dfe['Power'].str.replace('bhp','')

#there are 'null' also in these columns ,so we want to remove these too
dfe['Engine']=dfe['Engine'].str.replace('null','0')
dfe['Power']=dfe['Power'].str.replace('null','0')
dfe['Mileage']=dfe['Mileage'].str.replace('null','0')
dfe
```

```python
dfe.dtypes
#uint ==>unsighned int

#power,mileage,engine are still remain as object
#we want to convert object ==>float/int

```

```python
# Datatype conversion

#mileage ==> float
#Engine ==>float
#Power ==>float

dfe['Mileage']=dfe['Mileage'].astype(float)
dfe['Engine']=dfe['Engine'].astype(float)
dfe['Power']=dfe['Power'].astype(float)
```

```python
dfe.dtypes
```

```python
# we have done that 'null' ==>0 but actually no cars have 0 mileage,power and engine
#so we want to convert as missing value
# finally fill these missing value
```

```python
# convertion to missing value float ===> missing value
dfe.loc[dfe.Mileage==0,'Mileage']=np.NaN
dfe.loc[dfe.Power==0,'Power']=np.NaN
dfe.loc[dfe.Engine==0,'Engine']=np.NaN
```

```python
dfe.isna().sum()
```

```python
# fill missing values with
#mileage,engine,power ===>mean
#seat ==>mode
```

```python
dfe['Mileage']=dfe['Mileage'].fillna(dfe['Mileage'].mean())
dfe['Engine']=dfe['Engine'].fillna(dfe['Engine'].mean())
dfe['Power']=dfe['Power'].fillna(dfe['Power'].mean())
dfe['Seats']=dfe['Seats'].fillna(dfe['Seats'].mode()[0])
```

```python
dfe.isna().sum()
```

```python
dfe
```

```python
# price is in middle
# so drop price
#then seperate x&y
```

```python
x=dfe.drop(['Price'],axis=1)
y=dfe['Price']
```
