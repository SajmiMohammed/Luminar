```python
# encoding Objects
df1=pd.get_dummies(df[['Item_Fat_Content','Item_Type','Outlet_Size','Outlet_Identifier','Outlet_Location_Type']],drop_first=True)
df1
```

```python
dfe=pd.concat([df,df1],axis=1)
dfe
```

```python
dff=dfe.drop(['Item_Identifier','Item_Fat_Content','Item_Type','Outlet_Size','Outlet_Identifier','Outlet_Location_Type'],axis=1)
dff
```

```python
dff.dtypes
```

```python
dff.isna().sum()
```

```python
dff['Item_Weight']=dff['Item_Weight'].fillna(dff['Item_Weight'].mean())
```

```python
dff.isna().sum()
```

```python
x=dff.drop(['Outlet_Type'],axis=1)
y=dff['Outlet_Type']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=7)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
matr=confusion_matrix(y_pred,y_test)
print(matr)
```

```python
score=accuracy_score(y_pred,y_test)
score
```

```python
rep=classification_report(y_pred,y_test)
print(rep)
```
