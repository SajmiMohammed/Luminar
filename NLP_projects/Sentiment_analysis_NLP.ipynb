***Preprocessing of target column***

```python
# replace with positive=+1,negative=-1,neutral=0 ==>use map
df.replace({'Target':{'Positive':1,'Negative':-1,'Neutral':0}},inplace=True)
df
# df['Target]=df['Target].map({'Positive':1,'Negative':-1,'Neutral':0})
```

***Preprocessing of Text column***

```python
# Importing subpackages
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# assigning Text data to variable tweets
tweets=df.Text
tweets
```

```python
# Tokenization ( word by word data)=> TweetTokenizer
from nltk import TweetTokenizer
tk=TweetTokenizer()
tweets=tweets.apply(lambda x:tk.tokenize(x)).apply(lambda x:" ".join(x))
tweets
```

```python
# Removing special characters ==>apply regular exp
tweets=tweets.str.replace('[^a-zA-Z0-9]+',' ')          # pattern used in reg exp
tweets
```

```python
# Removing unmeaningful words other than stopwords
from nltk import word_tokenize
tweets=tweets.apply(lambda x:' '.join([w for w in word_tokenize(x) if len(w)>=3]))
tweets
```

```python
## Stemming =>SnowballStemmer [ finding root words]
from nltk.stem import SnowballStemmer
stemmer=SnowballStemmer('english')
tweets=tweets.apply(lambda x:[stemmer.stem(i.lower()) for i in tk.tokenize(x)]).apply(lambda x:' '.join(x))
tweets
```

```python
# Removing Stopwords
from nltk.corpus import stopwords
stop=stopwords.words('english')
tweets=tweets.apply(lambda x:[i for i in word_tokenize(x) if i not in stop]).apply(lambda x:' '.join(x))
tweets
```

```python
# Vectorization [word by word data to numerical data]
from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer()
train_data=vec.fit_transform(tweets)
print(train_data)
```

```python
# shape of train_data
train_data.shape
```

***Modelling***

```python
y=df['Target'].values
y
```

```python
x=train_data
```

```python
# Training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.30)
```

```python
# Model creation
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeRegressor
knn=KNeighborsClassifier()
model_nb=MultinomialNB()
model_svc=SVC()
model_tree=DecisionTreeRegressor()
lst=[knn,model_nb,model_svc,model_tree]
```

***Performance Evaluation***

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
for i in lst:
  print(i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print('-'*100)
  print("Accuray_score = ",accuracy_score(y_test,y_pred))
  print(' '*100)
  print(confusion_matrix(y_test,y_pred))
  print(' '*100)
  print(classification_report(y_test,y_pred))
```

***RESULT*** : We get higher accuracy score in SVC model.
